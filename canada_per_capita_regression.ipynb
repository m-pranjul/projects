##### Importing Necessary Libraries

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
```

##### Importing dataset file from drive

```python
# df = pd.read_csv("canada_per_capita_income.csv")
file_id = "1aX4iLpgbRQdYc2ZVg8j-q3YbzoFCfL5t"
df = pd.read_csv(f"https://drive.google.com/uc?id={file_id}")
df.head()
```

##### Analysing Data whether it is a good fit for linear regression or not??

```python
%matplotlib inline
plt.scatter(df["year"],df["per capita income (US$)"], color = "red", marker = "X")
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.title("Per Capita Income of Candian People")
plt.show()
```

##### Separating dependent and independent features from dataframe

```python
year = df.drop(["per capita income (US$)"], axis = "columns")
income = df["per capita income (US$)"]
```

```python
year.head()
```

```python
income.head(3)
```

##### Creating a Linear Model

```python
model = linear_model.LinearRegression()
```

##### Training the Linear Model

```python
model.fit(year, income)
```

##### Predicting Per Capita Income from linear model created for year 2020

```python
model.predict([[2020]])
```

```python
model.coef_
```

```python
model.intercept_
```
