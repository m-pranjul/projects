### Importing Digit images dataset

```python
%matplotlib inline

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

#### Splitting train and test samples

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(digits.data, digits.target, test_size = 0.2)
```

#### Importing and Training Logistic Regression Model.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=600)
```

```python
model.fit(train_x, train_y)
```

### .
#### Warning says failed to converge because of less number of iterations, so we give number of iterations,  
#### **max_iter = 600**.

```python
len(train_x)
```

```python
len(test_x)
```

```python
model.score(test_x, test_y)
```
